```python
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive

# Authenticate the client.
gauth = GoogleAuth()
gauth.LocalWebserverAuth()  # Creates local webserver and auto handles authentication.

drive = GoogleDrive(gauth)

# Function to create a folder
def create_folder(parent_id, folder_name):
    folder_metadata = {
        'title': folder_name,
        'mimeType': 'application/vnd.google-apps.folder',
        'parents': [{'id': parent_id}]
    }
    folder = drive.CreateFile(folder_metadata)
    folder.Upload()
    return folder['id']

# Create the main folder
deep_learning_folder_id = create_folder('root', 'Deep_Learning')

# List of weeks
weeks = ['week12', 'week13', 'week14']
subfolders = ['data', 'tutorials', 'exercises', 'solutions']

# Create folders for each week and subfolders within each week
for week in weeks:
    week_folder_id = create_folder(deep_learning_folder_id, week)
    for subfolder in subfolders:
        create_folder(week_folder_id, subfolder)

print("Folder structure created successfully.")
```
